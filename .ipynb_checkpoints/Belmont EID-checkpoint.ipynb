```python
import pandas as pd
import camelot
%matplotlib notebook
```

```python
# Include all pages and check how many tables there are on each page
belmont_tables = camelot.read_pdf("Belmont Equine Injury Database.pdf", flavor='lattice', pages='2')
belmont_tables
```

```python
# See what first table looks like
df = belmont_tables[3].df
df
```

```python
# Drop the first three rows
df = df.drop([0,1])
df
```

```python
# Rename all the columns
df = df.rename(columns={
    0: 'Year',
    1: 'Number of Starts',
    2: 'Number of Fatalities',
    3: 'Number of Fatalities per 1000 Starts'
})
df
```
